```python
import numpy as np
import pandas as pd
```

```python
new = pd.read_csv('polarity.csv')
```

```python
new['ID'] = new['ID'].astype(str)
new.info()
```

```python
def songs(pol):
    if -1.0 <= pol <= -0.5:
        rec = pd.read_csv('-10to-5.csv')
        print("You generally like songs with polarity ", pol)
        print("Here's a list recommended for you: ")
        df1 = rec[['Song Name','Artist Name']]
        print(df1)

    if -0.5 <= pol <= 0.00:
        rec = pd.read_csv('-5to0.csv')
        print("You generally like songs with polarity ", pol)
        print("Here's a list recommended for you: ")
        df1 = rec[['Song Name','Artist Name']]
        print(df1)

    if 0.0 <= pol <= 0.5:
        rec = pd.read_csv('0to5.csv')
        print("You generally like songs with polarity ", pol)
        print("Here's a list recommended for you: ")
        df1 = rec[['Song Name','Artist Name']]
        print(df1)

    if 0.5 <= pol <= 1.00:
        rec = pd.read_csv('5to10.csv')
        print("You generally like songs with polarity ", pol)
        print("Here's a list recommended for you: ")
        df1 = rec[['Song Name','Artist Name']]
        print(df1)
    return

while True:
    print()
    print("Recommendation on songs")
    print()
    print("MENU OPTIONS")
    print("1 - User-ID Based")
    print("2 - Polarity Based")
    print("0 - Exit")
    print()
    choice = input("> ")
    print()
    
    if choice == "0":
        break
    
    elif choice == "1":
        print("Enter User-ID")
        uid = input("> ")                
        pos = new.loc[new['ID'] == uid, 'polarity']
        print(pos[1])
        songs(pos[1])
            
    elif choice == "2":
        print("Enter Polarity")
        pol = input("> ") 
        songs(float(pol))
        
    else:
        print("Wrong input.\nTry again!")
            
```
